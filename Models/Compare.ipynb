```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.ticker import FuncFormatter

data = {
    "Model": ["LSTM", "Transformer", "Random Forest", "WLS"],
    "RMSE": [0.0002921, 0.0002388, 0.000938, None],
    "R² Score": [0.7122, 0.6923, 0.6017, 0.1759],
    "QLIKE": [0.0616, 0.0917, 0.2287, 0.2207]
}
df = pd.DataFrame(data)

def scientific_fmt(x, pos):
    return f"{x:.1e}"

fig, axs = plt.subplots(3, 1, figsize=(8, 12))

axs[0].bar(df["Model"], df["RMSE"], color="skyblue", edgecolor="black")
axs[0].set_title("Model Comparison: RMSE (lower is better)")
axs[0].set_ylabel("RMSE")
axs[0].yaxis.set_major_formatter(FuncFormatter(scientific_fmt))

axs[1].bar(df["Model"], df["R² Score"], color="lightgreen", edgecolor="black")
axs[1].set_title("Model Comparison: R² Score (higher is better)")
axs[1].set_ylabel("R²")

axs[2].bar(df["Model"], df["QLIKE"], color="salmon", edgecolor="black")
axs[2].set_title("Model Comparison: QLIKE (lower is better)")
axs[2].set_ylabel("QLIKE")

plt.tight_layout()
plt.savefig("model_comparison_chart.png")
plt.show()
```
